```python
import networkx as nx
```

## Создание графа

```python
G = nx.DiGraph()
edges = [
    (0, 1), (1, 2), (2, 1), (2, 3), (3, 2), (3, 4), (4, 3), (4, 5),
    (5, 6), (6, 7), (7, 8), (8, 9), (9, 6), (9, 11), (11, 9), (11, 10),(10, 11),
    (12, 10), (10, 12), (13, 12), (12, 13), (13, 14), (14, 13),
    (14, 15), (15, 16), (15, 17), (16, 17), (17, 18), (14, 18), (14, 19), (19, 18),
    (18, 0)
]
G.add_edges_from(edges)
# G.add_node(20)
nx.draw(G, with_labels=True);
```

## Матрицы смежности и инцидентности

```python
nx.adjacency_matrix(G).toarray()
```

```python
nx.incidence_matrix(G, oriented=True).toarray()
```

## Расстояния между вершинами

```python
paths = list(nx.all_pairs_shortest_path_length(G))
paths
```

## Эксцентриситет каждой вершины

```python
nx.eccentricity(G)
# у 20 вершины - 0
```

## Диаметр графа

```python
nx.diameter(G)
```

## Радиус графа

```python
nx.radius(G)
```

## Периферия графа

```python
nx.periphery(G)
```

## Центральные вершины

```python
list(
    filter(lambda x: x[1] == nx.radius(G), nx.eccentricity(G).items())
)
# вершины 13 и 14
```

## Центр графа

```python
nx.center(G)
```
