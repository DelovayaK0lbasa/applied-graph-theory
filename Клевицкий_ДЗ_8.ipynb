```python
import networkx as nx
```

```python
edges = [
    (1, 2, 5), (1, 4, 4),
    (2, 11, 3), (2, 12, 6), (2, 3, 9),
    (4, 3, 2),
    (11, 13, 12), (13, 8, 8), (8, 5, 9),
    (12, 5, 11),
    (3, 5, 4), (3, 6, 3), (6, 10, 7), (10, 9, 4), (9, 5, 9),
    (5, 7, 1), (5, 14, 3),
    (7, 15, 8), (15, 16, 12), (7, 16, 7), (7, 14, 10),
    (14, 17, 2), (14, 16, 5), (17, 16, 6)
]
G = nx.Graph()
G.add_weighted_edges_from(edges)
```

```python
veider_path = list(nx.eulerian_circuit(G, 1))
print(1, end='')
for path in veider_path:
    print(f' -> {path[1]}', end='')
```
